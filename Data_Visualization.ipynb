```python
def df_shifted(df, target=None, lag=0):
    if not lag and not target:
        return df       
    new = {}
    for c in df.columns:
        if c == target:
            new[c] = df[target]
        else:
            new[c] = df[c].shift(periods=lag)
    return  pd.DataFrame(data=new)
```

```python
df_1=df[['PriceUSD','Btc_Tweet_Count','commentsPerDay','Google_Trend']]
```

```python
df_new = df_shifted(df_1, 'Google_Trend', lag=-1)
```
